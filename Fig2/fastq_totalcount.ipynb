```python
import os
import csv

```

```python


# Specify the directory you want to start from
rootDir = '/work/users/s/h/shuang9/rip/rip_data'

results = []

#If you're certain about the standard 4-line format of your .fastq files 
# (i.e., each read consists of 4 lines: @header, sequence, +, quality), 
# a much faster approach would be to just count the total number of lines and divide by 4.

for dirName, subdirList, fileList in os.walk(rootDir):
    for fname in fileList:
        if fname.endswith('.fastq'):
            print(fname)
            with open(os.path.join(dirName, fname)) as f:
                count = sum(1 for _ in f)
                results.append([fname, count // 4])

# Write the results to a CSV file
with open('fastq_read_counts.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['FileName', 'TotalReads'])
    writer.writerows(results)


```
